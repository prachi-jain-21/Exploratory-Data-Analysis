# PyDataset

```python
# Import package
from pydataset import data

# importing numpy library
import pandas as pd

# importing packages
import pandas_profiling as pp
```

```python
# storing as data frame
pydataframe = pd.DataFrame(data())

# list of all datasets
print(pydataframe[['dataset_id']].to_string())
```

```python
# download datasets
df_aircraft = data('aircraft')

# as output.html file
profile = pp.ProfileReport(df_aircraft)
profile.to_file("output_aircraft.html")
```

```python
# download datasets
df_WorldPhones = data('WorldPhones')

# as output.html file
profile = pp.ProfileReport(df_WorldPhones)
profile.to_file("pandas_profiling_WorldPhones.html")
```
